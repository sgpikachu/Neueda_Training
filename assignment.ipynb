```python
import pandas as pd
import numpy as np

# Read data into a DataFrame
df_SPX = pd.read_csv('https://s3.eu-west-1.amazonaws.com/neueda.conygre.com/pydata/SPX.csv',
                     index_col='Date', parse_dates=True)

df_SPX
```

```python
col_list = ['Price','Open','High','Low']
for col in col_list:
    df_SPX[col]=pd.to_numeric(df_SPX[col].str.replace(',',''))
```

```python
df_SPX['Change %']=pd.to_numeric(df_SPX['Change %'].str.replace('%',''))
df_SPX
```

```python
df_SPX.sort_values('Date').reset_index(drop=True)
df_SPX
```

```python
import matplotlib.pyplot as plt
rows = "Date >= '2015-06-01' and Date<='2016-06-30'"
cols = ['High','Low']
df_draw = df_SPX.query(rows)[cols]
df_draw.plot.line()
```

```python
col_list = ['Price','Open','High','Low','Change %']
for col in col_list:
    dfmin = df_SPX[col].min()
    dfmax = df_SPX[col].max()
    dfavg = df_SPX[col].mean()
    print('Column', col, 'has the minimum of', dfmin, 'has the max of', dfmax, 'has the average of',dfavg)
```
